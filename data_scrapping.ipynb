```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r=requests.get("https://www.forbes.com/lists/india-billionaires/")
print(r)
soup=BeautifulSoup(r.text,"html.parser")


table=soup.find_all("section",class_="csf-row et-rows table-container row_5kmknec6y5oni")

```

```python
span=soup.find_all('span',class_="header-content-text")
header_texts = []
for s in span:
  text=s.text
  header_texts.append(text)

print(header_texts)
```

```python
df=pd.DataFrame(columns=header_texts)
print(df)
```

```python
data=soup.find_all("div",class_="row-cell-value")

for d in data:
  num=d.text
  print(num)

```

```python
all_data = []

for d in data:
    num = d.text
    all_data.append(num)

print(all_data)
```

```python
# Reshape the all_data list into a list of lists, where each inner list is a row
reshaped_data = [all_data[i:i+4] for i in range(0, len(all_data), 4)]

# Create the DataFrame using the reshaped data and the header texts as columns
df2 = pd.DataFrame(reshaped_data, columns=header_texts)
df2.index=df2.index+1
display(df2)
```

```python
df2.to_csv("billionaires.csv")
```
